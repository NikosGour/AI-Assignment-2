```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/project2_dataset.csv')
```

```python
df.head()
```

```python
df.drop(columns=['Month', 'Browser', 'OperatingSystems'], inplace=True)
```

```python
cols = df.columns[df.dtypes == 'bool']
df[cols] = df[cols].astype(int)
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['Region', 'TrafficType', 'VisitorType'])
```

```python
df.describe()
```

```python
df.info()
```

```python
from sklearn.preprocessing import MinMaxScaler

min_max = MinMaxScaler()
scaling_columns = df.columns[(df.dtypes == 'int64') | (df.dtypes == 'float64')]
min_max = min_max.fit(df[scaling_columns])
df[scaling_columns] = min_max.transform(df[scaling_columns])
```
